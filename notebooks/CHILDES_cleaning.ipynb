## Plotting Cleanups

### Parenthesis

```python
%matplotlib inline
from rich.console import Console
import collections
import matplotlib.pyplot as plt
from lexical_benchmark import settings
from lexical_benchmark.datasets import utils

root_dir = settings.PATH.raw_childes

with Console().status("Counting Words.."):
    words = []
    for bfile in (root_dir / "Eng-NA/child").glob("*.par"):
        words.extend([s.strip() for s in bfile.read_text().splitlines()])
    wc = collections.Counter(words)
    sorted_words = sorted(wc.items(), key=lambda x: x[1], reverse=True)
    print(f"{len(sorted_words)=}")
```

```python
words, frequencies = zip(*sorted_words, strict=True)
plt.figure(figsize=(10, 6))
plt.bar(words, frequencies)
plt.xlabel("Word")
plt.ylabel("Frequency")
plt.title("Word Frequency Distribution (ALL)")
plt.xticks(rotation=90)
plt.savefig('img/words_frequency_par_first50.svg', bbox_inches='tight', pad_inches=1, format='svg')
```

### Brackets

```python
%matplotlib inline
from rich.console import Console
import collections
import matplotlib.pyplot as plt
from lexical_benchmark import settings
from lexical_benchmark.datasets import utils

root_dir = settings.PATH.raw_childes

with Console().status("Counting Words.."):
    words = []
    for bfile in (root_dir / "Eng-NA/child").glob("*.brackets"):
        words.extend([s.strip() for s in bfile.read_text().splitlines()])
    wc = collections.Counter(words)
    
with Console().status("Sorting.."):
    sorted_words = sorted(wc.items(), key=lambda x: x[1], reverse=True)
```

```python
len(sorted_words)
```

```python
sorted_words2 = sorted_words[:50]
words, frequencies = zip(*sorted_words2, strict=True)
plt.figure(figsize=(10, 6))
plt.bar(words, frequencies)
plt.xlabel("Word")
plt.ylabel("Frequency")
plt.title("Word Frequency Distribution(First 50)")
plt.xticks(rotation=90)
plt.savefig('img/words_frequency_brackets_first_50.svg', bbox_inches='tight', pad_inches=1, format='svg')
```

```python
sorted_words2 = sorted_words[-50:]
words, frequencies = zip(*sorted_words2, strict=True)
plt.figure(figsize=(10, 6))
plt.bar(words, frequencies)
plt.xlabel("Word")
plt.ylabel("Frequency")
plt.title("Word Frequency Distribution(Last 50)")
plt.xticks(rotation=90)
plt.savefig('img/words_frequency_brackets_last_50.svg', bbox_inches='tight', pad_inches=1, format='svg')
```

```python
import random
sample_size = 0.005
print(int(len(sorted_words) * sample_size))
sorted_words2 = list(random.sample(sorted_words, int(len(sorted_words) * sample_size)))

words, frequencies = zip(*sorted_words2, strict=True)
plt.figure(figsize=(10, 6))
plt.bar(words, frequencies)
plt.xlabel("Word")
plt.ylabel("Frequency")
plt.title("Word Frequency Distribution(Sample 50)")
plt.xticks(rotation=90)
plt.savefig('img/words_frequency_brackets_sample_50_3.svg', bbox_inches='tight', pad_inches=1, format='svg')
```
